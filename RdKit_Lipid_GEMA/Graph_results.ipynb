```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
```

## iBD1106

```python

final_results = r"final_results/iBD1106_last.xlsx"

df = pd.read_excel(final_results)
df      
```

```python
  
# plotting graph for lipid class representation
lc=df.iloc[0:15].plot(x="Lipid Class", y=["TP","FN", "FP"], kind="bar",color=("green","orange","red"),width=0.8,figsize=(8, 5))

# Set y-axis scale and limit
lc.set_ylim(0, 35)  
for p in lc.patches:
    lc.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height() * 1.04),ha='center', va='bottom', rotation=90)
# Show the plot

plt.show()

# plotting graph for metrics
m_df =df[["Lipid Class","Precision","Recall","F1"]]
m_df.Recall=m_df.Recall.round(3)
m_df.F1=m_df.F1.round(3)

mt=m_df.plot(x="Lipid Class", y=["Precision", "Recall","F1"], kind="bar",color=("mediumorchid","darksalmon","gold"),width=0.8,figsize=(8, 5))

# Set y-axis scale and limit
mt.set_ylim(0, 1.1)  
for p in mt.patches:
    mt.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height() * 1.04),ha='center', va='bottom', rotation=90)
# Show the plot
plt.show()

```

## Radar Graph

```python
import numpy as np
from matplotlib.patches import Patch

# Create lists with the global metrics for each model
lithologies = ["Precision","Recall","F1"]
iNS937 = [1,0.101,0.183]
iLB1027_lipid = [1,0.859,0.924]
iCZ843_H = [1,0.434599,0.605882]
iBD1106 = [1,0.443299,0.614286]

# to perform a polygon we need to close the list in a circle, with the same first and last element
lithologies = [*lithologies, lithologies[0]]
iLB1027_lipid = [*iLB1027_lipid, iLB1027_lipid[0]]
iCZ843_H = [*iCZ843_H, iCZ843_H[0]]
iNS937 = [*iNS937, iNS937[0]]
iBD1106 = [*iBD1106, iBD1106[0]]

# creating the plot
label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(lithologies))

fig, ax = plt.subplots(figsize=(10,10), subplot_kw=dict(polar=True))

ax.plot(label_loc, iLB1027_lipid, lw=2,c='C0')
ax.plot(label_loc, iCZ843_H, lw=2,c='black')
ax.plot(label_loc, iNS937, lw=2,c='C1')
ax.plot(label_loc, iBD1106, lw=2,c='C3')

ax.fill(label_loc, iLB1027_lipid,alpha=0.3,c='C0')
ax.fill(label_loc, iCZ843_H, alpha=0.3,c='black')
ax.fill(label_loc, iNS937, alpha=0.3,c='C1')
ax.fill(label_loc, iBD1106, alpha=0.3,c='C3')

#Create legend
iLB1027_lipid_legend = Patch(facecolor='C0', alpha=0.5, label='iLB1027_lipid')
iCZ843_H_legend = Patch(facecolor='black', alpha=0.5, label='iCZ843_H')
iNS937_legend = Patch(facecolor='C1', alpha=0.5, label='iNS937')
iBD1106_legend = Patch(facecolor='C3', alpha=0.5, label='iBD1106')

ax.legend(handles=[iBD1106_legend, iLB1027_lipid_legend,iNS937_legend,iCZ843_H_legend],
              bbox_to_anchor=(1, 0.2), fontsize=20, 
              frameon=True)

lines, labels = plt.thetagrids(np.degrees(label_loc), labels=lithologies)
ax.tick_params(axis='both', which='major', pad=20, labelsize=13)

plt.show()
```
