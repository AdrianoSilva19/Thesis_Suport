## Radar Graph

```python
import numpy as np
from matplotlib.patches import Patch

# Create lists with the global metrics for each model
lithologies = ["Precision","Recall","F1"]
iNS937 = [1,0.101,0.183]
iLB1027_lipid = [1,0.859,0.924]
iCZ843_H = [1,0.434599,0.605882]
iBD1106 = [1,0.443299,0.614286]

# to perform a polygon we need to close the list in a circle, with the same first and last element
lithologies = [*lithologies, lithologies[0]]
iLB1027_lipid = [*iLB1027_lipid, iLB1027_lipid[0]]
iCZ843_H = [*iCZ843_H, iCZ843_H[0]]
iNS937 = [*iNS937, iNS937[0]]
iBD1106 = [*iBD1106, iBD1106[0]]

# creating the plot
label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(lithologies))

fig, ax = plt.subplots(figsize=(8,7), subplot_kw=dict(polar=True))

ax.plot(label_loc, iLB1027_lipid, lw=2,c='C0')
ax.plot(label_loc, iCZ843_H, lw=2,c='black')
ax.plot(label_loc, iNS937, lw=2,c='C1')
ax.plot(label_loc, iBD1106, lw=2,c='C3')

ax.fill(label_loc, iLB1027_lipid,alpha=0.3,c='C0')
ax.fill(label_loc, iCZ843_H, alpha=0.3,c='black')
ax.fill(label_loc, iNS937, alpha=0.3,c='C1')
ax.fill(label_loc, iBD1106, alpha=0.3,c='C3')

#Create legend
iLB1027_lipid_legend = Patch(facecolor='C0', alpha=0.5, label='iLB1027_lipid')
iCZ843_H_legend = Patch(facecolor='black', alpha=0.5, label='iCZ843_H')
iNS937_legend = Patch(facecolor='C1', alpha=0.5, label='iNS937')
iBD1106_legend = Patch(facecolor='C3', alpha=0.5, label='iBD1106')

ax.legend(handles=[iBD1106_legend, iLB1027_lipid_legend,iNS937_legend,iCZ843_H_legend],
              bbox_to_anchor=(1, 0.2), fontsize=13, 
              frameon=True)

lines, labels = plt.thetagrids(np.degrees(label_loc), labels=lithologies)
ax.tick_params(axis='both', which='major', pad=20, labelsize=13)

plt.show()
```

```python
query =" MATCH (p:Compound) WHERE exists(p.swiss_lipids_id) and exists(p.lipidmaps_id) RETURN p.boimmg_id as boimmg_id"
from tqdm import tqdm

def query_database(query):
    uri="bolt://palsson.di.uminho.pt:6091"
    user="neo4j"
    password="1234"
    counter = []
    driver = GraphDatabase.driver(uri, auth=basic_auth(user, password))
    with driver.session() as session1:
        node = session1.run(query)
        for value in node:
            counter.append (value.get("boimmg_id"))
    return counter

def get_parents(ont_id: int) -> list:
        """Get predecessors using as parameter the database identifier and the relationship type"""
        uri="bolt://palsson.di.uminho.pt:6091"
        user="neo4j"
        password="1234"
        driver = GraphDatabase.driver(uri, auth=basic_auth(user, password))
        parents = []
        relationship_type = "is_a"
        with driver.session() as session:
            result = session.run(
                "MATCH (c:Compound)-[r]->(p:Compound) "
                "WHERE c.boimmg_id = $ont_id AND TYPE(r) = $rel_type "
                "RETURN p.name as name ",
                ont_id=ont_id,
                rel_type=relationship_type,
            )

            data = result.data()
            if data:
                for node in data:
                    name = node.get("name")
                    parents.append(
                       name
                    )
        return parents

list_results = query_database(query)
parents = []
for value in tqdm(list_results):
    parents.append(get_parents(value))
print(parents)
```

```python
component_counts = {}

for sublist in parents:
    for component in sublist:
        if component in component_counts:
            component_counts[component] += 1
        else:
            component_counts[component] = 1

sorted_dict = dict(sorted(component_counts.items(), key=lambda item: item[1],reverse=True))

print(sorted_dict)

# Threshold for grouping into "Others"
threshold = 200


labels = []
sizes = []

others = 0
for key, value in sorted_dict.items():
    if value >= threshold:
        labels.append(key)
        sizes.append(value)
    else:
        others += value

if others > 0:
    labels.append('Others')
    sizes.append(others)

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)


# Display the chart
plt.axis('equal')
plt.show()
```
